```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Transfer Function
numerator = [1, 1]
denominator = [1, 4, 7, 6, 3]
G = ctrl.TransferFunction(numerator, denominator)
G
```

```python
# Poles and Zeros
poles = ctrl.poles(G)
zeros = ctrl.zeros(G)
print('Poles:', poles)
print('Zeros:', zeros)
```

```python
# Step Response
t, y = ctrl.step_response(G)
plt.figure(figsize=(10,5))
plt.plot(t, y, color='#0D9488', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Pole-Zero Map
plt.figure(figsize=(6,6))
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()
```
