# Control Systems Analysis using SymPy and Python Control
This notebook derives the transfer function from a differential equation and analyzes the system using Python.

## 1. Import Libraries

```python
%matplotlib inline
import sympy as sp
from control import tf, step_response, poles, zeros, pzmap
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 6]
```

## 2. Define Laplace Variables

```python
s = sp.symbols('s')
X = sp.symbols('X')
Y = sp.symbols('Y')
```

## 3. Convert Differential Equation to Laplace Form
Equation:
2 d²y/dt² + dy/dt = dx/dt + x

```python
A = 2*s**2 * Y + s * Y
B = s * X + X
```

## 4. Solve for Transfer Function G(s) = Y/X

```python
tf_expr = sp.solve(sp.Eq(A, B), Y)[0] / X
tf_simplified = sp.simplify(tf_expr)
print('Transfer Function G(s):', tf_simplified)
```

## 5. Convert to Control System Format

```python
num_coeffs = [float(c) for c in sp.Poly(sp.numer(tf_simplified), s).all_coeffs()]
den_coeffs = [float(c) for c in sp.Poly(sp.denom(tf_simplified), s).all_coeffs()]
sys = tf(num_coeffs, den_coeffs)
print(sys)
```

## 6. Poles and Zeros

```python
print('Poles:', poles(sys))
print('Zeros:', zeros(sys))
```

## 7. Step Response

```python
t, y = step_response(sys, T=np.linspace(0, 20, 1000))
plt.figure(figsize=(10, 6))
plt.plot(t, y, 'b-', linewidth=2)
plt.title('Step Response', fontsize=14)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 8. Pole-Zero Map

```python
plt.figure(figsize=(10, 8))
pzmap(sys, grid=True)
plt.title('Pole-Zero Map', fontsize=14)
plt.tight_layout()
plt.show()
```
