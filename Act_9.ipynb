# Control Systems Analysis using SymPy and Python Control
This notebook derives the transfer function from a differential equation and analyzes the system using Python.

## 1. Import Libraries

```python
import sympy as sp
from control import tf, step_response, poles, zeros, pzmap
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Define Laplace Variables

```python
s = sp.symbols('s')
X = sp.symbols('X')
Y = sp.symbols('Y')
```

## 3. Convert Differential Equation to Laplace Form
Equation:
2 d²y/dt² + dy/dt = dx/dt + x

```python
A = 2*s**2 * Y + s * Y
B = s * X + X
```

## 4. Solve for Transfer Function G(s) = Y/X

```python
tf_expr = sp.solve(sp.Eq(A, B), Y)[0] / X
tf_simplified = sp.simplify(tf_expr)
print('Transfer Function G(s):', tf_simplified)
```

## 5. Convert to Control System Format

```python
num_coeffs = [float(c) for c in sp.Poly(sp.numer(tf_simplified), s).all_coeffs()]
den_coeffs = [float(c) for c in sp.Poly(sp.denom(tf_simplified), s).all_coeffs()]
sys = tf(num_coeffs, den_coeffs)
print(sys)
```

## 6. Poles and Zeros

```python
print('Poles:', poles(sys))
print('Zeros:', zeros(sys))
```

## 7. Step Response

```python
t, y = step_response(sys)
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## 8. Pole-Zero Map

```python
pzmap(sys, grid=True)
plt.show()
```
