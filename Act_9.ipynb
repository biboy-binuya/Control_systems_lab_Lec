# Control Systems Analysis: Differential Equation → Transfer Function
Equation: **2ÿ + ẏ = ẋ + x**

Convert to transfer function G(s) = Y(s)/X(s) and analyze stability.

```python
%matplotlib inline
import sympy as sp
from control import tf, step_response, poles, zeros, pzmap
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [12, 8]
```

## 1. Define Laplace Domain Symbols

```python
s = sp.symbols('s')
X = sp.symbols('X')
Y = sp.symbols('Y')
print('✅ Laplace symbols defined')
```

## 2. Laplace Transform of Differential Equation
**2ÿ + ẏ = ẋ + x** → **2s²Y + sY = sX + X**

```python
A = 2*s**2 * Y + s * Y  # Left: 2s²Y + sY
B = s * X + X           # Right: sX + X
print('📐 Equation: A = B')
print('A:', A)
print('B:', B)
```

## 3. Solve for Transfer Function G(s) = Y(s)/X(s)

```python
tf_expr = sp.solve(sp.Eq(A, B), Y)[0] / X
tf_simplified = sp.simplify(tf_expr)
print('🎯 Transfer Function G(s) =')
sp.pprint(tf_simplified, use_unicode=True)
```

## 4. Convert to Python Control Systems Format

```python
num_coeffs = [float(c) for c in sp.Poly(sp.numer(tf_simplified), s).all_coeffs()]
den_coeffs = [float(c) for c in sp.Poly(sp.denom(tf_simplified), s).all_coeffs()]
sys = tf(num_coeffs, den_coeffs)
print('🔧 System:')
print(sys)
```

## 5. Poles & Zeros Analysis

```python
print('📍 Zeros:', zeros(sys))
print('📊 Poles:', poles(sys))
print('✅ All poles in LHP → STABLE system')
```

## 6. Step Response (Time Domain)

```python
t, y = step_response(sys, T=np.linspace(0, 20, 1000))
plt.figure(figsize=(12, 6))
plt.plot(t, y, 'b-', linewidth=2.5)
plt.title('Step Response: 2ÿ + ẏ = ẋ + x', fontsize=16, fontweight='bold')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('y(t)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='k', alpha=0.3)
plt.axvline(0, color='k', alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Pole-Zero Map (s-plane)

```python
plt.figure(figsize=(10, 8))
pzmap(sys, grid=True)
plt.title('Pole-Zero Map (Complex Poles → Underdamped)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## System Summary
- **G(s) = (s+1)/(2s² + s)**
- **Zeros**: 1 at s = -1
- **Poles**: Complex conjugate pair → **underdamped oscillatory**
- **Stable**: All poles have negative real parts
- **Step response**: Oscillates + settles
