```python
import control as ctrl
import matplotlib.pyplot as plt
# Numerator and denominator coefficients
num = [1, 1]
den = [1, 4, 7, 6, 3]
# Create the transfer function
G = ctrl.tf(num, den)

# Print G
print("Transfer Function G:")
print(G)
Transfer Function G:
<TransferFunction>: sys[7]
Inputs (1): ['u[0]']
Outputs (1): ['y[0]']

              s + 1
  -----------------------------
  s^4 + 4 s^3 + 7 s^2 + 6 s + 3
# Compute and print poles and zeros
poles = G.poles()
zeros = G.zeros()
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
Poles: [-1.5+0.8660254j -1.5-0.8660254j -0.5+0.8660254j -0.5-0.8660254j]
Zeros: [-1.+0.j]
# Generate a pole-zero map
plt.figure()
ctrl.pzmap(G, grid=True)
plt.title('Pole-Zero Map')
plt.show()

# Plot the system step response
plt.figure()
t, y = ctrl.step_response(G)
plt.plot(t, y)
plt.title('System Step Response')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
